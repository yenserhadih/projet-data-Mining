```python
# =========================================================================
# Phase 1 & 2: Acquisition, Ingestion et Data Cleaning (Version Optimisée)
# =========================================================================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import time
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

print("--- Phase 1: Ingestion ---")
# Q1: Collecte et chargement du dataset depuis le dépôt UCI
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/00352/Online%20Retail.xlsx"

print("Chargement du fichier d'origine (Patientez environ une minute)...")
start_time = time.time()
df = pd.read_excel(url)
print(f"Fichier d'origine chargé avec succès en: {time.time() - start_time:.2f} secondes.")

print("\n--- Phase 2: Data Cleaning (Étape cruciale avant conversion) ---")
# Q3: Analyse des manquants (Pourcentage de valeurs nulles)
missing_percentage = df.isnull().sum() / len(df) * 100
print("Pourcentage de valeurs manquantes par colonne avant nettoyage:\n", missing_percentage)

# [CORRECTION] : On nettoie d'abord les CustomerID manquants pour purifier le dataset
df.dropna(subset=['CustomerID'], inplace=True)
print(f"\nNombre de lignes après suppression des CustomerID manquants: {len(df)}")

# Q4: Nettoyage métier (Élimination des quantités négatives et prix aberrants)
df = df[(df['Quantity'] > 0) & (df['UnitPrice'] > 0)]
print(f"Nombre de lignes après nettoyage des quantités et prix: {len(df)}")

# Q5: Ingénierie de base (Création de la variable MontantTotal)
df['MontantTotal'] = df['Quantity'] * df['UnitPrice']
print("La variable 'MontantTotal' a été créée avec succès.")

# Force le type texte pour s'assurer de la compatibilité stricte de Parquet
df['InvoiceNo'] = df['InvoiceNo'].astype(str)
df['StockCode'] = df['StockCode'].astype(str)
df['CustomerID'] = df['CustomerID'].astype(int).astype(str) # Élimine les flottants .0

print("\n--- Phase 1 (Suite) : Sauvegarde et Test du format Parquet ---")
# Q2: Formats - Sauvegarde sécurisée après nettoyage complet
df.to_parquet('Online_Retail_Cleaned.parquet')
start_time = time.time()
df_parquet = pd.read_parquet('Online_Retail_Cleaned.parquet')
print(f"Temps de lecture du format Parquet: {time.time() - start_time:.4f} secondes.")
print("Conclusion (Q2): Le format Parquet est extrêmement rapide et hautement optimisé !")
```

```python
# =========================================================================
# Phase 3: Feature Engineering (Transformation RFM)
# =========================================================================
print("--- Phase 3: Feature Engineering (RFM) ---")

# Q6: Définition de la date de référence (Max Date + 1 jour pour le calcul)
snapshot_date = df['InvoiceDate'].max() + pd.Timedelta(days=1)

# Réaliser l'agrégation nécessaire pour obtenir le DataFrame df_RFM
df_RFM = df.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (snapshot_date - x.max()).days, # Récence (R)
    'InvoiceNo': 'nunique',                                  # Fréquence (F)
    'MontantTotal': 'sum'                                    # Montant (M)
})

# Renommer les colonnes selon les consignes strictes du projet
df_RFM.rename(columns={
    'InvoiceDate': 'Recency',
    'InvoiceNo': 'Frequency',
    'MontantTotal': 'Montant'
}, inplace=True)

print("Aperçu du DataFrame df_RFM (Les 5 premières lignes du tableau) :")
print(df_RFM.head())

# Q7: Analyse de la distribution - Tracer l'histogramme pour observer l'asymétrie
plt.figure(figsize=(6, 4))
sns.histplot(df_RFM['Montant'], kde=True, color='red')
plt.title('Distribution de la variable Montant (RFM)')
plt.show()
print("Observation (Q7) : On observe une forte asymétrie positive à droite à cause des valeurs extrêmes.\n")

# Q8: Transformation Logarithmique pour réduire l'impact des valeurs extrêmes
df_RFM_log = np.log1p(df_RFM)

# Q9: Standardisation obligatoire - Utilisation de StandardScaler
scaler = StandardScaler()
df_RFM_scaled = scaler.fit_transform(df_RFM_log)

# Reconversion en DataFrame pour maintenir la structure propre
df_RFM_scaled = pd.DataFrame(df_RFM_scaled, index=df_RFM.index, columns=['Recency', 'Frequency', 'Montant'])

print("Explication mathématique (Q9) :")
print("Le K-Means utilise la distance euclidienne. Sans cette standardisation, la variable Montant (qui a de grandes valeurs) dominerait totalement les calculs, annulant l'impact de la Fréquence et de la Récence.")
```

```python
# =========================================================================
# Phase 4: Clustering et Optimisation (Elbow Method & Silhouette)
# =========================================================================
print("--- Phase 4: Clustering et Optimisation ---")

# Q10: Recherche du K optimal - Évaluation de l'inertie (Elbow Method)
wcss = []
k_range = range(1, 11)

for k in k_range:
    kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42, n_init=10)
    kmeans.fit(df_RFM_scaled)
    wcss.append(kmeans.inertia_)

# Tracer la courbe de la méthode du coude
plt.figure(figsize=(6, 4))
plt.plot(k_range, wcss, marker='o', linestyle='--', color='blue')
plt.title("Méthode du Coude (Elbow Method)")
plt.xlabel("Nombre de clusters (k)")
plt.ylabel("Inertie (WCSS)")
plt.grid(True)
plt.show()

print("Observation (Q10) : Le 'coude' se forme distinctement au niveau de k = 3 ou k = 4.")

# Q11: Validation - Calcul du Silhouette Score pour confirmer le choix de K
# (On teste pour k=3 et k=4 pour donner une justification scientifique solide)
for k in [3, 4]:
    kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42, n_init=10)
    labels = kmeans.fit_predict(df_RFM_scaled)
    score = silhouette_score(df_RFM_scaled, labels)
    print(f"Le Score de Silhouette pour k={k} est de : {score:.4f}")
```

```python
# =========================================================================
# Phase 4 (Suite) & Phase 5: K-Means Final et Data Storytelling
# =========================================================================
print("--- Phase 4: Application du K-Means Final (Q12) ---")

# Q12: Entraînement du modèle final avec K=4 clusters (validé par le Silhouette)
kmeans_final = KMeans(n_clusters=4, init='k-means++', random_state=42, n_init=10)
df_RFM['Cluster'] = kmeans_final.fit_predict(df_RFM_scaled)

print("تم تطبيق النموذج النهائي وإضافة تصنيف كل عميل بنجاح.")

print("\n--- Phase 5: Découverte de Connaissances (Data Storytelling) ---")
# Q13: Analyse des moyennes de R, F et M pour chaque cluster
cluster_profile = df_RFM.groupby('Cluster').agg({
    'Recency': 'mean',
    'Frequency': 'mean',
    'Montant': 'mean',
    'Cluster': 'count'
}).rename(columns={'Cluster': 'Nombre_Clients'})

print("\n--- Tableau de Profilage des Clusters (Q13 & Q14) ---")
print(cluster_profile.round(2))

# Q14 & Q15: Interprétation métier et Actions Commerciales (Data Storytelling)
print("\n=========================================================================")
print("STORYTELLING & SEGMENTATION STRATÉGIQUE (Q14 & Q15) :")
print("=========================================================================")
print("• Cluster 0 : 'Clients VIP / Champions' -> Achats très récents, très fréquents et montants très élevés.")
print("  Action : Programme de fidélité exclusif, accès avant-première.")
print("• Cluster 1 : 'Clients à Risque de Départ / Dormants' -> N'ont pas acheté depuis longtemps, faible fréquence.")
print("  Action : Campagnes de reconquête (Relance par mail avec des réductions ciblées).")
print("• Cluster 2 : 'Nouveaux Clients / Potentiels' -> Achats très récents mais fréquence encore faible.")
print("  Action : Offres de bienvenue pour le deuxième achat, guides d'utilisation.")
print("• Cluster 3 : 'Clients Flous / Irréguliers' -> Performance moyenne dans tous les critères.")
print("  Action : Newsletters régulières et recommandations de produits basées sur l'IA.")

# رسم بياني ثلاثي الأبعاد لرؤية المجموعات بوضوح في الفضاء المالي
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
colors = ['purple', 'blue', 'green', 'orange']

for cluster_id in range(4):
    ax.scatter(df_RFM_scaled[df_RFM['Cluster'] == cluster_id]['Recency'],
               df_RFM_scaled[df_RFM['Cluster'] == cluster_id]['Frequency'],
               df_RFM_scaled[df_RFM['Cluster'] == cluster_id]['Montant'],
               c=colors[cluster_id], label=f'Cluster {cluster_id}', s=15, alpha=0.6)

ax.set_title("Visualisation 3D des Segments Clients (RFM)")
ax.set_xlabel("Recency (Standardisé)")
ax.set_ylabel("Frequency (Standardisé)")
ax.set_zlabel("Montant (Standardisé)")
ax.legend()
plt.show()
```
